```python
uni_towns = open('university_towns.txt')
uni_towns.readlines()[:20]
```

```python
uni_towns = []
with open('university_towns.txt') as file:
    for line in file:
        if '[edit]' in line:
            state = line
        else:
            uni_towns.append((state, line))
uni_towns[:20]
```

```python
uni_towns_df = pd.DataFrame(uni_towns, columns=["State", "RegionName"])
uni_towns_df
```

```python
def get_citystate(item):
    if ' (' in item:
        return item[:item.find(' (')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
```

```python
get_citystate('Wyoming[edit]')
```

```python
get_citystate('Laramie (University of Wyoming)[5]\n')
```

```python
uni_towns_df = uni_towns_df.applymap(get_citystate)
uni_towns_df
```
